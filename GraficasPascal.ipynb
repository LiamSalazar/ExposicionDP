```python
import time
import matplotlib.pyplot as plt

def C_rec(n, k):
    if k == 0 or k == n:
        return 1
    return C_rec(n-1, k-1) + C_rec(n-1, k)

def triangle_rec(n):
    tri = []
    for i in range(n+1):
        row = []
        for j in range(i+1):
            row.append(C_rec(i, j))
        tri.append(row)
    return tri

def triangle_dp(n):
    tri = []
    for i in range(n+1):
        row = [1] * (i+1)
        for j in range(1, i):
            row[j] = tri[i-1][j-1] + tri[i-1][j]
        tri.append(row)
    return tri

ns = list(range(0, 16))
times_rec = []
times_dp = []

for n in ns:
    start = time.perf_counter()
    triangle_dp(n)
    times_dp.append(time.perf_counter() - start)
    
    start = time.perf_counter()
    triangle_rec(n)
    times_rec.append(time.perf_counter() - start)


plt.figure(figsize=(10, 5))
plt.plot(ns, times_dp, label="DP (O(n^2))")
plt.plot(ns, times_rec, label="Recursivo (O(2^n))")
plt.xlabel("n")
plt.ylabel("Tiempo (s)")
plt.title("Comparaci√≥n de tiempos: DP vs Recursivo")
plt.legend()
plt.grid(True)
plt.show()
```
